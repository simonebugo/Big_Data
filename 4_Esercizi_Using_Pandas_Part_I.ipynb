```
import pandas as pd
import numpy as np
```

# Load data

```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url, header=None, names=['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class'])

#metodo .head() di un DataFrame Pandas restituisce le prime 5 righe del dataset per default
iris.head()
```

# Data cleaning

## Missing values

### Is there any missing value in the dataframe?

```
pd.isnull(iris).sum()
```

### Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```
iris.iloc[10:30,2:3] = np.nan #selezioni co√¨ la riga dalla 10 alla 29 e la colonna di indice 2 e le setto a nan
iris.head(30)
```

### Which column has the maximum number of missing values?

```
missing_per_columns = pd.isnull(iris).sum() #per ogni colonna conto il numero di valori nulli
missing_per_columns.index[missing_per_columns.argmax()]
```

```
missing_per_columns.idxmax() #altro modo non utilizzando l'argmax ma usando direttamente idxmax
```

### Try to substitute the NaN values with two methods:
- replace null values with column mean (apply it to a copy of the dataframe)
- replace null values with 1.0



```
new_iris = iris.copy()
new_iris = new_iris.fillna(iris.loc[:,["sepal_length","sepal_width","petal_length","petal_width"]].mean())
new_iris.head(20)
```

```
new_iris = iris.copy()
new_iris = new_iris.fillna(1.0)
new_iris.head(20)
```

### Set the first 3 rows as NaN

```
iris.iloc[0:3 ,:] = np.nan
iris.head()
```

### Delete the rows that have all NaN

```
iris = iris.dropna(how='all')
iris.head(20)
```

### Reset the index so it begins with 0 again

```
iris = iris.reset_index(drop = True) #siccome avevamo droppato tutte le righe che contenevano tutti Nan allora il dataset aveva gli indici sbagliati. cos√¨
#metto nuovamente gli indici da 0 in poi
iris.head(20)
```

## Duplicates

### Does the dataframe contain duplicated rows? If any, visualize all duplicated rows (don't omit first or last occurrences)

```
iris_duplicate_rows = iris[iris.duplicated(keep=False)] #cos√¨ mostra tutte le righe duplicate, non solo quali sono le righe che venogno luplicate come sarebbe con:
#iris_duplicate_rows = iris[iris.duplicated(keep='first')]
iris_duplicate_rows
```

### Which row is the most repeated?

```
iris_duplicate_rows.groupby(list(iris_duplicate_rows.columns.values)).size()
```

### Drop duplicated rows

```
iris = iris.drop_duplicates()
```

## Detect outliers, e.g., values that are higher than 85th percentile and lower than 25th percentile.

```
iris_copy = iris.loc[:,"sepal_length":"petal_width"].copy()
iris_copy[(iris_copy > iris_copy.quantile(.85)).all(axis=1)]
iris_copy[(iris_copy < iris_copy.quantile(.25)).all(axis=1)]
```

# Data transformation

## Replace class values by removing "Iris-" prefix (use a dictionary)

```
iris.head()
```

```
class_map_values = {}
for class_val in iris['class'].unique():
  class_map_values[class_val] = class_val.replace("Iris-", "")
class_map_values
```

```
iris['class'] = iris['class'].replace(class_map_values)
iris.head()
```

## Delete columns
Delete for example class column

```
del iris['class']
iris
```

## How to normalize all columns in a dataframe?
- Normalize all columns of df by subtracting the column mean and divide by standard deviation.
- Range all columns of df such that the minimum value in each column is 0 and max is 1.

```
# Solution Q1
out1 = iris.apply(lambda x: ((x - x.mean())/x.std()).round(2)) #semplicemente appico la formula richiessta.
# round(2) arrotonda i valori numerici a 2 cifre decimali.
print('Solution Q1\n',out1)

# Solution Q2
out2 = iris.apply(lambda x: ((x.max() - x)/(x.max() - x.min())).round(2)) #riesco cos' a limitare il valore di una colonna tra 0 e 1
print('Solution Q2\n', out2)
```

## Binning and discretization
Discretize dataframe columns in 4 bins and get the new value frequency distribution

```
discrete_iris = iris.apply(lambda x: pd.cut(x,4)) #prende una colonna x e la suddivide in 4 intervalli (bin) con pd.cut.
discrete_iris
```

```
discrete_iris.sepal_length.unique()
```

```
discrete_iris.sepal_length.value_counts(normalize=True)
```
