### Standardization, or mean removal and variance scaling


Standardization of datasets is a common requirement for many machine learning estimators implemented in scikit-learn;

In practice we often ignore the shape of the distribution and just transform the data to center it by removing the mean value of each feature, then scale it by dividing non-constant features by their standard deviation.



in sklearn è possibile applicare delle tecniche per modificare i valori delle feature. abbiamo preso i valori delle 4 feature e abbiamo usato quei valori.  avremmo potuto applicare delle tecniche di scaling per modificare i valori e vedere se ha impatto sui risultati del ml. alcuni algoritmi funzionano meglio con valori scalati per es in un intervallo [0,1].
in sklearn ci sono diversi trasformatori che possono essere usati tutti allo stesso modo. hanno una funzione che si chiama fit che impara qualcosa dai dati e calcola delle funzione statistiche sui dati. es media deviazione standard ecc... e poi funzione trasform che permette di applicare trasformazione ai dati. poi c'è fit_trasform che fa entrambe le cose. prima fa la fit e poi trasforma i dati. vediamo alcune di qeuste possibil trasformazioni

```python
#prima funzione: lo scaling.
from sklearn import preprocessing
import numpy as np
import pandas as pd

X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])

X_scaled = preprocessing.scale(X_train)   # default axis=0 standardize each feature

X_scaled
```

The standard score of a sample x is calculated as:
<tt>z = (x - u) / s</tt>
where u is the mean of the training samples or zero if <tt>with_mean=False</tt>, and s is the standard deviation of the training samples or one if <tt>with_std=False</tt>.


The preprocessing module further provides a utility class StandardScaler that implements the Transformer API to compute the mean and standard deviation on a training set so as to be able to later reapply the same transformation on the testing set.

standard scaler che applica una funzione ai dati per cui per ogni feature calcola media e deviazione standard. dopo ad ogni valore delle feature si sottrae
la media e si divide per la deviazione standard. i valori vengono trasformati in moco che ogni feature abbia media 0 e deviazione standard 1. alcun algoritmi fuznionano meglio con i valori a cui è applicato lo standard scaler.

```python
sss = preprocessing.StandardScaler() #inizializzo standard scaler
scaler = sss.fit(X_train) #fit calcola medie e deviazione standard su traing
#una volta inizalizzato il trasformatore e applicata funzione di fit il trasformatore si può applicare su qualsiasi altro dato (anche sui dati di test)
#media e dev stanrd che userà sono quelle imparate dai dati di train
```

```python
scaler.transform(X_train) #trasform va a modificare effettivamente i dati in x_train.
```

```python
scaler.transform(X_train).mean(), scaler.transform(X_train).std() #controllo media e dev stanrd di ogni feature scopro che c'è una mediaa di 0 e una dev stardard di 1
```

The scaler instance can then be used on new data to transform it the same way it did on the training set



```python
X_test = np.array([[ 2., -3.,  2.],
                    [ 2.,  5.,  0.],
                    [ 0.,  1., -1.]])
```

```python
scaler.transform(X_test)
```

```python
scaler.fit(X_train)
scaler.transform(X_train)
```

```python
scaler.fit_transform(X_train)
```

### Scaling features to a range

An alternative standardization is scaling features to lie between a given minimum and maximum value, often between zero and one, or so that the maximum absolute value of each feature is scaled to unit size.

```python
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
min_max_scaler = preprocessing.MinMaxScaler((0,1)) # default [0,1]
#min max scaler alrto tipo di trasformatore che applcia una trasforamzione per fare in mdoo che ogni feature abbai valori compresi tra 0 e 1. e ho feature con intervallo + alto questi valori
#vengono ritorri nell'intervallo [0,1] mantenendo i rapporti tra i vari valori. intervallo 0,1 è quello di default ma si può peronalizzare
```

```python
min_max_scaler.fit(X_train)
```

```python
min_max_scaler.transform(X_train)
```

```python
X_test = np.array([[-3., -1.,  4.]])
min_max_scaler.transform(X_test) #come visto prima una volta che ho creato il trasformatore, posso chiamare il trasformatore su un altro set di dati diverso da quello visto durante
#il fit del trasformatore
```

The transformation is given by:

X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

X_scaled = X_std * (max - min) + min

If your data contains many outliers, scaling using the mean and variance of the data is likely to not work very well. In these cases, you can use robust_scale and RobustScaler (x_scaled = x / max(abs(x)))
as drop-in replacements instead. They use more robust estimates for the center and range of your data.



```python
from sklearn.preprocessing import RobustScaler #anzi che usare media e deviazione standard, usa la mediana e l'interquartirle (differenza tra terzo quartile e primo quartile.)
X = [[ 1., -2.,  2.],
     [ -2.,  1.,  3.],
     [ 4.,  1., -2.]]
```

```python
transformer = RobustScaler().fit(X)
```

```python
transformer.scale_
```

```python
transformer.transform(X)
```

The transformation is computed as:

<tt>X_scaled = (X - X.median(axis=0)) / X.IQR(axis=0)</tt>

The IQR is the range between the 1st quartile (q1) and the 3rd quartile (q3).

## Normalization


Normalization is the process of scaling individual samples to have unit norm. This process can be useful if you plan to use a quadratic form such as the dot-product or any other kernel to quantify the similarity of any pair of samples.

‘l1’
The l1 norm uses the sum of all the values and thus gives equal penalty to all parameters, enforcing sparsity.
x_normalized = x / sum(X)


‘l2’
The l2 norm uses the square root of the sum of all the squared values. This creates smoothness and rotational invariance. Some models, like PCA, assume rotational invariance, and so l2 will perform better.
x_normalized = x / sqrt(sum((i**2) for i in X))

normalizzazione fa in modo che la norma di ogni riga del dataset sia uguale a 1. la somma di tutti i quadrati se la calcoliamo è uguale a 1. altra tipoloia di trasformazione.

```python
X = np.array([[1., -1.,  2.],
              [ 2.,  0.,  0.],
              [ 0.,  1., -1.]])
X_normalized = preprocessing.normalize(X, norm='l2')
```

## Discretization

discretizzazione: permette di fare quanto già visto con la funzione cut. con cut potevamo splittare, dividere i valori di un intervallo continuo in tanti intervalli. discretizziamo una variabile.
stessa cosa si fa con KBinsdiscretizer. si può dare in input il numero di gruppi in cui deve essere diviso un insieme di valori. i valori verranno poi trasformati secondo diverse strategie. ci sono diverse strateige.
uniform divide i valori in intervalli tutti della stessa ampiezza. ho un intervallo che va da 0 a 20 e voglio dividerli in 4 bins si creano 4 gruppi in cui il primo va da 0-5,5-10 ec... tutti gruppi della stesssa ampiezza
quantile divide questi valori in mdoo che in ogni gruppo ci sia lo stesso numero di sample. crea lui gli estremi di ogni intervallo.

Discretization (otherwise known as quantization or binning) provides a way to partition **continuous features into discrete values**. Certain datasets with continuous features may benefit from discretization, because discretization can transform the dataset of continuous attributes to one with only nominal attributes.

```python
X = np.array([[ -3., 5., 15 ],
              [  0., 6., 14 ],
              [  6., 3., 11 ]])
est = preprocessing.KBinsDiscretizer(n_bins=[3, 2, 3], encode='ordinal').fit(X)
est
```

```python
est.transform(X)
```

There are different strategies implemented in KBinsDiscretizer:

- ‘uniform’: The discretization is uniform in each feature, which means that the bin widths are constant in each dimension.

- ‘quantile’: The discretization is done on the quantiled values, which means that each bin has approximately the same number of samples.

- ‘kmeans’: The discretization is based on the centroids of a KMeans clustering procedure.

https://scikit-learn.org/stable/auto_examples/preprocessing/plot_discretization_strategies.html?highlight=kbinsdiscretizer

Feature binarization is the process of thresholding numerical features to get boolean values. This can be useful for downstream probabilistic estimators that make assumption that the input data is distributed according to a multi-variate Bernoulli distribution.

## Custom transformers

Often, you will want to convert an existing Python function into a transformer to assist in data cleaning or processing.

abbiamo visto come utilizzare trasformatori già implementati. esiste function transformer. mi permette di decideere qual'è la funzione di trasformazione e usare le facility di questo trasformer, ovvero suare fit e trasform.

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
#tipicamente i trasformatori di sklearn danno in output un array numpy. applicati su array numpy o dataframe e danno in output array numpy/dataframe

transformer = FunctionTransformer(np.log1p, validate=True) #funzione data in input è il log1p. funzione calcola il logaritmo di 1+x, funzione applicata a tutti i valori diun certo array numpy e dataframe
#oltre alla funzione la function trasformer prende in input il validate. valide = TRUE significa che il dato viene prima validato e poi trasformato. validato significa che viene trasforamto in arry numpy
#se do in input un arraynumpy al trasforamtore con validate = true, x viene validato e l'output sarà un array numpy
X = np.array([[0, 1], [2, 3]])
transformer.transform(X)
```

The *validate* parameter indicates that the input X array should be checked before calling func. The possibilities are:

- If False, there is no input validation.

- If True, then X will be converted to a 2-dimensional NumPy array or sparse matrix. If the conversion is not possible an exception is raised.

```python
type(transformer.transform(X)) #il type di ciò che viene trasformato è ancora un arraynumpy
```
